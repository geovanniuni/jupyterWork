```python
###Ejercicio1
### ----- Ejercicio # 1 -----###
import numpy as np

def sumaElementos(v):
    s=0
    for i in range(len(v)):
        s = s + v[i]
    return s


def ordenarIndicesColumnas(m):
    men=999
    indices = []
    
    for j in range(len(m[0])):
    
        for i in range(len(m[0])):
            s=sumaElementos(m[:,i])
            if(s<men):
                men=s
                ix=i
        indices.append(ix)
        men=999
        m[:,ix] = np.array([999,999,999,999])
        
    return indices
        

def ordenarColumnas(m,indices):
    mr=m.copy()
    for i in range(len(indices)):
        mr[:,i]=m[:,indices[i]]
    return mr
    
    

lx = [[3,5,8,4,1],[1,2,7,3,1],[3,10,5,1,1],[6,7,8,2,1]]
m1 = np.array(lx)
m2 = m1.copy()
print(m1)

print("--------Matrix inicial----------")
print(len(m1))

indices=ordenarIndicesColumnas(m1)
print(indices)

print("-------------------------")

mx=ordenarColumnas(m2,indices)
print("------Matrix final--------------")
print(mx)
```

```python
import numpy as np
import matplotlib.pyplot as plt

rad = np.pi/180

def fx(x):
    y=np.sin(x)
    return y


def primeraDerivada(x):
    h=0.01
    y = (fx(x+h) - fx(x))/h
    return y

def obteniendoY(ang):
    ly=[]
    for i in range(len(ang)):
        y=fx(i*rad)
        ly.append(y)
    return ly
    

ang = np.arange(0,720,0.5)
y=obteniendoY(ang)
y= np.array(y)

y1=primeraDerivada(ang*rad)


plt.plot(ang,y,ang,y1)
plt.show()


```
